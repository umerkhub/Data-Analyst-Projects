## Dropping Duplicates

```python
sales_d = sales[sales.duplicated()]
sales_d
```

```python
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'])
sales_sort = sales.sort_values(by = 'Order_Date', ascending=True)
sales_sort
```

```python
sales = sales_sort.drop_duplicates(subset='Order_ID', keep='last')
sales
```

## Handling Null Values

```python
sales.isnull().sum()
```

```python
sales.fillna({'Discount':0}, inplace=True)
```

```python
sales.isnull().sum()
```

## Calculate IQR

### For `Unit_Price`

```python
unit_Q1 = sales['Unit_Price'].quantile(0.25)
unit_Q3 = sales['Unit_Price'].quantile(0.75)
unit_IQR = unit_Q3 - unit_Q1
display(f"Q1: {unit_Q1}, Q3: {unit_Q3}, IQR: {unit_IQR}")
```

```python
unit_lower_bound = unit_Q1 - 1.5 * unit_IQR
unit_upper_bound = unit_Q3 + 1.5 * unit_IQR
display(f"Lower Bound: {unit_lower_bound}, Upper Bound: {unit_upper_bound}")
```

```python
unit_outliers = sales[(sales['Unit_Price'] < unit_lower_bound) | (sales['Unit_Price'] > unit_upper_bound)]
display(unit_outliers)
```

```python
sales['Unit_Price'] = sales['Unit_Price'].clip(lower = unit_lower_bound, upper = unit_upper_bound)
display(sales)
```

### For `Quantity`

```python
qty_Q1 = sales['Quantity'].quantile(0.25)
qty_Q3 = sales['Quantity'].quantile(0.75)
qty_IQR = qty_Q3 - qty_Q1
display(f"Q1: {qty_Q1}, Q3: {qty_Q3}, IQR: {qty_IQR}")
```

```python
qty_lower_bound = qty_Q1 - 1.5 * qty_IQR
qty_upper_bound = qty_Q3 + 1.5 * qty_IQR
display(f"Lower Bound: {qty_lower_bound}, Upper Bound: {qty_upper_bound}")
```

```python
qty_outliers = sales[(sales['Quantity'] < qty_lower_bound) | (sales['Quantity'] > qty_upper_bound)]
display(qty_outliers)
```

```python
sales['Quantity'] = sales['Quantity'].clip(lower = qty_lower_bound, upper = qty_upper_bound)
display(sales)
```

## Handling Missing Values in `Total_Sales`

```python
sales['Total_Sales'] = sales['Unit_Price'] * sales['Quantity'] * (1 - sales['Discount'])
sales
```
